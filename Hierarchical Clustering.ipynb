# Hierarchical clustering algorithm

Hierarchical clustering is an umsupervised machine learning method that you can use to predict subgroups based on the difference between data points and their nearest neighbors.

Each data point is linked to its nearest neighbor based on the distance matrix you choose.

Imagine that you are a geneticist and want to identify functional gene groups by analyzing gene expression profiles or profiles that identify all genes active in a cell.

You decide to apply hierarchical clustering to identify these groups based on the distance between data points in the gene expression profile and their closest neighbors.

Hierarchical clustering predicts subgroups within data by finding the distance between each data point. and its nearest neighbor and also linking up the most nearby neighbors.

You can find the number of subgroups suitable for a hierarchical clustering model by looking at a dendrogram.



**Importing the Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings("ignore")
```

**Creating a Synthetic Dataset**

```python
example_data = make_blobs(n_samples = 300, n_features = 2,
                          centers = 3, cluster_std = 1.3, random_state = 42)
```

```python
example_data
```

```python
dots = example_data[0]
```

```python
dots
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram_example = sch.dendrogram(sch.linkage(dots, method = "ward"))
```

```python
dots[:,0]
```

```python
dots[:,1]
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(dots[:,0], dots[:,1])
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_clustering = AgglomerativeClustering(n_clusters=5, affinity = "euclidean", linkage = "ward")
```

```python
y_hier_clus = hierarchical_clustering.fit_predict(dots)
```

```python
y_hier_clus
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(dots[y_hier_clus == 0,0], dots[y_hier_clus == 0,1], s = 150, c = "blue")
plt.scatter(dots[y_hier_clus == 1,0], dots[y_hier_clus == 1,1], s = 150, c = "yellow")
plt.scatter(dots[y_hier_clus == 2,0], dots[y_hier_clus == 2,1], s = 150, c = "red")
plt.scatter(dots[y_hier_clus == 3,0], dots[y_hier_clus == 3,1], s = 150, c = "cyan")
plt.scatter(dots[y_hier_clus == 4,0], dots[y_hier_clus == 4,1], s = 150, c = "black")
plt.show()
```
