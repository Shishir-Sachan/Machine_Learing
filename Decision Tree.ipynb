# Decision Tree

### Max_depth
It represents the maximum depth of the decision tree.

If no value is entered, it will be unlimited. If the model is overfitting, it should be downgraded.

### Min_samples_split
This is the minimum number of samples a node should have before splitting.

### Min_samples_leaf
This would refer to the minimum number of samples a leaf should have.


### Max_features
This is the number of features to consider when looking for the best split.


# Decision Tree 

`Advantages`

It is easy to understand and interpret.

It actually requires very little data preparation However, I do want to make a noted that this model does not support missing values.

The cost of the tree is logarithmic to the number of data poits used points used to train the tree.

It can process both numerical and cayegorical data.

It can also handle problems with multiple outputs. In other words, data sets with more than one label number of the target variable can be studied with in this algorithm.

It is also very possible to verify a model using statistical tests.

`Disadvantages`

Sometimes, you can get some overly complex trees that they just need to explain the data better.

Pruning refers to remoning leaf nodes containing few objects from the decision tree graph.

```python
sns.pairplot(df, hue="Species", height = 2.75);
```

```python
df["Species_Num"] = df["Species"]
```

```python
df["Species_Num"] = df["Species_Num"].str.replace("setosa", "1")
df["Species_Num"] = df["Species_Num"].str.replace("versicolor", "2")
df["Species_Num"] = df["Species_Num"].str.replace("virginica", "3")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Species_Num"] = pd.to_numeric(df["Species_Num"], downcast = "integer")
```

```python
df.info()
```

```python
plt.figure(figsize = (15,9))
sns.heatmap(df.corr(), cmap='Spectral', annot = True);
```

```python
df.drop(["PetalLengthCm", "Species_Num"], axis = 1, inplace = True)
```

```python
df.head()
```

**Decision Tree Modelling**

```python
X = df.drop(["Species"], axis = 1)
y = df.Species
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores = cross_val_score(estimator=dt_model, X = X_train, y = y_train, cv=10)
cv_scores.mean()
```

```python
cv_scores
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cnf_matrix, annot=True, cmap="YlGnBu")
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
print(classification_report(y_test, y_pred))
```

**Hyperparameter Optimization with Python**

```python
from sklearn.model_selection import GridSearchCV
```

```python
dt_model2 = DecisionTreeClassifier()
```

```python
dt_params = {"criterion": ["gini", "entropy"],
              "max_depth" : np.arange(2, 14, 2),
              'max_features': ["auto", "sqrt", "log2"]}
```

```python
dt_cv_model = GridSearchCV(estimator = dt_model2, param_grid = dt_params, cv = 10)
```

```python
dt_cv_model.fit(X_train, y_train)
```

```python
dt_cv_model.best_params_
```

```python
dt_tuned = DecisionTreeClassifier(criterion = "gini", max_depth = 6, max_features = "sqrt")
```

```python
dt_tuned.fit(X_train, y_train)
```

```python
y_pred = dt_tuned.predict(X_test)
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
cv_scores = cross_val_score(estimator = dt_tuned, X = X_train, y = y_train, cv=10)
cv_scores.mean()
```

```python
plt.figure(figsize=(20,20)) 
tree2 = tree.plot_tree(dt_tuned, filled=True, feature_names=features , class_names = targets)
```
